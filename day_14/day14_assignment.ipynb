### Exercise 1

### Explain the difference between map, filter, and reduce.

#### Map:

Purpose: Applies a function to each element in an iterable and returns a new iterable of the transformed elements.
Syntax: map(function, iterable)
Example: Doubling each element in a list: map(lambda x: 2*x, [2, 4, 6]) returns [4, 8, 12].

```python
my_list = [2, 4, 6]
doubled_list = map(lambda x: 2*x, my_list)

print(list(doubled_list))
```

#### Filter:

Purpose: Filters elements from an iterable based on a condition given by a function and returns a new iterable containing only the elements that satisfy the condition.
Syntax: filter(function, iterable)
Example: Filtering even numbers from a list: filter(lambda x: x%2 == 0, [2, 3, 4, 5]) returns [2, 4].

```python
my_list = [2, 3, 4, 5]
even_list = filter(lambda x: x % 2 == 0, my_list)

print(list(even_list))
```

#### Reduce:

Purpose: Accumulates elements in an iterable by applying a function to pairs of elements successively. It starts with the first two elements, then applies the function to the result and the third element, and so on, until a single value remains.
Syntax: reduce(function, iterable, initial=None)
Example: Summing all elements in a list: reduce(lambda x, y: x+y, [1, 2, 3]) returns 6.

```python
my_list = [1, 2, 3]
sum_of_list = reduce(lambda x, y: x+y, my_list)

print(sum_of_list)
```

### Exercise 2

### Explain the difference between higher order function, closure and decorator

#### 1. Higher-order functions:

Definition: Functions that take other functions as arguments, return functions, or both. They essentially manipulate functions as data.
Example: map(function, iterable) takes a function and applies it to each element in an iterable.
Benefit: Enable code reuse, abstraction, and concise syntax for common functional patterns.

2. Closures:

Definition: Functions that "remember" and access variables from their enclosing scope, even after the enclosing function has finished execution.
Example: A counter function that keeps track of its calls even when it's no longer in the local scope.
Benefit: Allow stateful functions within stateless environments like pure functions.

3. Decorators:

Definition: Syntactic sugar for wrapping a function with another function to add functionality or modify its behavior without directly altering the original code.
Example: Adding timing or logging functionality to a function with a simple @decorator syntax.
Benefit: Enhance code readability, modularity, and separation of concerns.

### Exercise 3

### Define a call function before map, filter or reduce, see examples.

#### Defining a call function before map:

This is useful when you have a complex transformation logic that you want to reuse across multiple map calls. 

```python
def double(x):
    return 2 * x
my_list = [2, 4, 6]
doubled_list = list(map(double, my_list))

print(doubled_list)
```

#### Defining a call function before filter:

This is helpful when you have a specific condition you want to check repeatedly in different filter calls.

```python
def is_even(x):
    return x % 2 == 0
my_list = [1, 2, 3, 4]
even_list = list(filter(is_even, my_list))

print(even_list)
```

#### Defining a call function before reduce:

This is useful when you have a custom operation to apply to pairs of elements in the iterable.

```python
def sum_of_squares(x, y):
    return x**2 + y**2
my_list = [1, 2, 3]
total_squares = reduce(sum_of_squares, my_list)

print(total_squares)
```

```python
#Create a function returning a dictionary, where keys stand for starting letters of countries and values are the number of country names starting with that letter.
def count_countries_by_starting_letter(countries):
    starting_letter_count = {}

    for country in countries:
        starting_letter = country[0].upper()  # Assuming you want to count by uppercase starting letters
        starting_letter_count[starting_letter] = starting_letter_count.get(starting_letter, 0) + 1

    return starting_letter_count

result = count_countries_by_starting_letter(countries)

for starting_letter, count in result.items():
    print(f'{starting_letter}: {count} countries')
```
